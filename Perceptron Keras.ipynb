```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Random dataset

```python
n_pts = 500
np.random.seed(0)
Xa = np.array([np.random.normal(13, 2, n_pts),
               np.random.normal(12, 2, n_pts)]).T
Xb = np.array([np.random.normal(8, 2, n_pts),
               np.random.normal(6, 2, n_pts)]).T

X = np.vstack((Xa, Xb))
y = np.matrix(np.append(np.zeros(n_pts), np.ones(n_pts))).T

plt.scatter(X[:n_pts,0], X[:n_pts,1])
plt.scatter(X[n_pts:,0], X[n_pts:,1])
```

## Perceptron

Perceptron is Machine Learning algorithm for supervised learning of various binary classification tasks. Further, Perceptron is also understood as an Artificial Neuron or neural network unit that helps to detect certain input data computations in business intelligence.


### Importing Keras

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

Sequential model is a linear stack of layers.
Here we will have only input and output layers.

Dense class is used to connect every node to preceding layer.

Adam is an optimization algorithm.

Gradient descent calculates the error of all the data points every time to take a small step to calculate minima.
Adam is a sthocastic gradient descent algorithm. 

```python
model = Sequential()
```

```python
model.add(Dense(units=1, input_shape=(2,), activation='sigmoid'))
```

units = 1 means only 1 output

input shape is 2 as only 2 weights w1 and w2 are used

```python
adam = Adam(learning_rate=0.1)
model.compile(adam, loss='binary_crossentropy', metrics=['accuracy'])
```

Initially a learning rate of 0.1 is given. Adam will automatically adjust the learning rate.

The loss functions used is cross entropy given below:


```python
history = model.fit(x=X, y=y, verbose=1, batch_size=50, epochs=200, shuffle='true')
```

Verbosity or verbose means how many lines of log we want after each epoch.

1 epoch means when the entire dataset is read by the perceptron. Epochs should neither be too less nor too much.

Batch size determines the size of the batches in which 1 epoch is divided.

shuffle means that the rows will be shuffled and a subset of rows will be used so that the alogirthm might not get stuck in a local minima rather than finding the global minima.

```python
history.history.keys()
```

```python
plt.plot(history.history['accuracy'])
plt.title('accuracy')
plt.xlabel('epoch')
```

```python
plt.plot(history.history['loss'])
plt.title('loss')
plt.xlabel('epoch')
```

```python
def plot_decision_boundary(X, y, model):
    x_span = np.linspace(min(X[:,0]) - 1, max(X[:,0]) + 1)
    y_span = np.linspace(min(X[:,1]) - 1, max(X[:,1]) + 1)
    xx, yy = np.meshgrid(x_span, y_span)
    xx_, yy_ = xx.ravel(), yy.ravel()
    grid = np.c_[xx_, yy_]
    pred_func = model.predict(grid)
    z = pred_func.reshape(xx.shape)
    plt.contourf(xx, yy, z)
    

plot_decision_boundary(X, y, model)
plt.scatter(X[:n_pts,0], X[:n_pts,1])
plt.scatter(X[n_pts:,0], X[n_pts:,1])
```
